## Test Chat Completion

````python
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))

response = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[{"role": "user", "content": "Explain Retrieval-Augmented Generation in 3 bullets."}]
)
response.choices[0].message.content
````

## Embedding Similarity Test

````python
import numpy as np

sentences = [
    "Artificial intelligence is transforming industries.",
    "Machine learning is changing business operations."
]

emb = client.embeddings.create(model="text-embedding-3-small", input=sentences)
vec1 = np.array(emb.data[0].embedding)
vec2 = np.array(emb.data[1].embedding)

similarity = np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
similarity
````
